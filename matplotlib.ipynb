```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv('gas_prices.csv')
print(gas)
```

```python
plt.title('Valor do combustível em U$')

list_country = ['Australia', 'Italy', 'USA', 'Japan', 'Mexico', 'France', 'Germany']

for country in gas:
    if country in list_country:
        plt.plot(gas['Year'], gas[country],label = [country], marker='.')

plt.xticks(gas['Year'][::2])


plt.xlabel('Years')
plt.ylabel('U$')
plt.legend()
plt.savefig('Country.png')
plt.show()
```
